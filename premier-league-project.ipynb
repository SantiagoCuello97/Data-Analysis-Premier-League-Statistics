## Exploratory Analysis and Visualization




Let's begin by importing`matplotlib.pyplot` and `seaborn`.

```python
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline

sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (36, 20)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
```

 #       <p style="text-align: center;">CHAMPIONSHIPS AND RELEGATIONS</p>
 **First of all, let's see which teams have won the Premier League in this century.**

```python
winners = (epl_df[epl_df.Pos == 1])
winners
```

```python
most_times_w = winners.Team.value_counts()
```

```python
sns.catplot(y = 'Team', kind='count', data = winners, aspect = 2.5, order = most_times_w.index, palette = 'summer')
plt.title('Champions between 2000-2022', fontsize = 20)
plt.xlabel('Titles')
```

```python
best_campaings = winners.sort_values('Pts',ascending=False)
best_campaings.head(1)
```

#### **We can see Manchester City is the best champion of the century winning the league in season 2017-18 with 100 points. Let's see now teams that ended in the last positions.**

```python
losers = (epl_df[epl_df.Pos >= 18])
relegations = losers.Team.value_counts()
```

```python
plt.figure(figsize=(12, 12))
sns.barplot(x=relegations, y=relegations.index, palette = 'autumn')
plt.title('Relegations between 2000-2022');
plt.xlabel('Relegations');
plt.ylabel('Team');
```

 **West Bromwich Albion and Norwich City are the teams that have had more relegations from 2000 to 2022.**

```python
worse_campaings = losers.sort_values('Pts',ascending=False)
worse_campaings.tail(1)
```

**Derby County has had so far the worst campaign of the century getting just 11 points in the 2007-08 season.**

```python
points_df = epl_df.groupby('Team').agg({'Pts': 'sum'})
matches_df = epl_df.groupby('Team').agg({'Pld':'sum'})
points_df.sort_values('Pts', ascending=False)
matches_df.sort_values('Pld', ascending=False)

effectivity_df = pd.merge(points_df, matches_df, on='Team')
effectivity_df["Total Points"] = effectivity_df['Pld']*3
effectivity_df["Eff"] = effectivity_df['Pts']/effectivity_df['Total Points']


eff_df = effectivity_df.copy()
del eff_df['Pts']
del eff_df['Pld']
del eff_df['Total Points']

eff_df = eff_df.sort_values("Eff", ascending = True)


te = eff_df.plot(kind='barh',figsize=(12,14), color='xkcd:sky blue')
plt.title('Team Effectiveness between 2000-2022',fontsize=20)
plt.ylabel('Team',fontsize=20)
plt.xlabel('Effectivity',fontsize=20);
te.set_xlim(0.0, 1)
```

**This was calculated considering points/games played for each team in all seasons from 2000 to 2022. We can see Manchester United is leading with a 67,7% effectivity. Let's now analyze the Top 5 Effective teams in the XXI Century.**

 #       <p style="text-align: center;">MANCHESTER UNITED</p>

![Cover](https://images6.alphacoders.com/969/thumb-1920-969527.jpg)

```python
united_df = epl_df[epl_df.Team == 'Manchester United']
united_df

united_df.plot(x='Season', y='Pos', marker = 'o', ms = '6', lw = '3', mfc = 'black', c = 'red', alpha = 0.8, figsize=(15,10))

plt.xlabel('Season')
plt.ylabel('Position')

plt.title("Manchester United positions between 2000-2022", fontsize = 20)
plt.legend(['Pos']);
plt.gca().invert_yaxis()
```

```python
united_goals_df = united_df.groupby('Season').agg({'GF': 'sum'})

axun = united_goals_df.plot(kind='bar',figsize=(15,10), color='red' , alpha = 0.8)
plt.xticks(rotation=65)
plt.title('Manchester United total goals per season between 2000-2022',fontsize=20)
plt.ylabel('Goals',fontsize=20)
plt.xlabel('Season',fontsize=20);
axun.set_ylim(20, 110)
```

```python
wunited_df = united_df.groupby('Team').agg({'W': 'sum'})
lunited_df = united_df.groupby('Team').agg({'L': 'sum'})
dunited_df = united_df.groupby('Team').agg({'D': 'sum'})
tplayed_united_df = united_df.groupby('Team').agg({'Pld': 'sum'})

wlunited_df = pd.merge(wunited_df, lunited_df, on='Team')
wldunited_df = pd.merge(wlunited_df, dunited_df, on='Team')

stats_united_df = pd.merge(wldunited_df, tplayed_united_df, on='Team')

pie_united_df = stats_united_df.copy()

pie_united_df['W'] = pie_united_df['W']/pie_united_df['Pld']
pie_united_df['L'] = pie_united_df['L']/pie_united_df['Pld']
pie_united_df['D'] = pie_united_df['D']/pie_united_df['Pld']

del pie_united_df['Pld']

#colors = ['red', 'yellowgreen', 'grey']
colors = ["#2ca02c", "#d62728", "grey"]


at = pie_united_df.iloc[0].plot.pie(autopct='%1.1f%%', colors = colors, explode=[0.01, 0.01, 0.01], shadow=True,  startangle=0, legend=True, title='Air Termination System', ylabel='', labeldistance=None)
plt.title("Manchester United\n" + "Percentage of wins, draws and losses between 2000 - 2022", bbox={'facecolor':'0.9', 'pad':10}, fontsize = 17)
```

On the contrary, on the second decade, althought  its level dropped significantly, being reflected both in the number of goals (scoring less than 50 goals in the 2015-2016 season) and its positions in the table where it was several times off the podium.

2) Manchester City's situation is the opposite of its eternal rival Manchester United. We can see that in the first decade of the century the club had a hard time in the Premier League, being relegated in the first season and always staying in mid-table positions in the following years, scoring few goals. But as the second decade began, the level of the team increased a lot, finishing 11 times on the podium, of which they won the championship 6 times. The same can be seen reflected in the number of goals scored where it passed 6 times the 90-goal mark and 3 times the 100-goal mark.

3) Chelsea, the second most efficient team of the century so far, has had fairly good and regular years in the Premier League these two decades, finishing 14 times on the podium and winning 5 titles. This regularity can also be seen reflected in its goals, where the average is around 70.

4) Considering the history and importance that the club has, Liverpool has not had a good start of the century. Looking at the data we can see that during these 22 years the club managed to finish 10 times on the podium, but it crowned only once. This led to leave behind the years of being the most champion of England, giving the crown to Manchester United.

6) Derby County has had so far the worst campaign of the century getting just 11 points in the 2007-08 season.

8) The Premier League saw its best season of the century in 2018-19, where the total points scored was 1068.
